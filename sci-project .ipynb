```python
data = {
    "Year": [2020, 2021, 2022, 2023, 2024, 2025],
    "Aswan_Visitors": [5000, 122000, 200000, 300000, 400000, 1400000], 
    "Egypt_Visitors": [3500000, 8000000, 11700000, 14900000, 15700000, 18800000],  
    "Luxor_Visitors": [525000,1200000, 1755000, 2235000, 2355000, 2820000]        
}

df = pd.DataFrame(data)
```

```python
df["Aswan_percent"]=df["Aswan_Visitors"]/df["Egypt_Visitors"] * 100
```

```python
plt.figure(figsize=(12,6))
bar_width = 0.35
years = df["Year"]
plt.bar(years - 0.17, df["Egypt_Visitors"], width=bar_width, color='lightgray', label="Egypt Total")
plt.bar(years + 0.17, df["Aswan_Visitors"], width=bar_width, color='skyblue', label="Aswan")
plt.title("Comparison of Visitors: Aswan vs Egypt (2020-2025)")
plt.xlabel("Year")
plt.ylabel("Number of Visitors")
plt.xticks(years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
plt.figure(figsize=(6,6))
sizes = [df["Aswan_Visitors"].iloc[-1], df["Egypt_Visitors"].iloc[-1] - df["Aswan_Visitors"].iloc[-1]]
labels = ["Aswan", "Rest of Egypt"]
colors = ['skyblue', 'lightgray']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.1,0))
plt.title("Aswan's Share of Egypt's Tourism (2025)")
plt.show()
```

```python
df["Luxor_Percent"] = df["Luxor_Visitors"] / df["Egypt_Visitors"] * 100
```

```python
plt.figure(figsize=(12,6))
bar_width = 0.30
years = df["Year"]
x = np.arange(len(years))  

plt.bar(x - bar_width, df["Egypt_Visitors"],
        width=bar_width, color='lightgray', label="Egypt Total")
plt.bar(x, df["Luxor_Visitors"],
        width=bar_width, color='orange', label="Luxor")
plt.bar(x + bar_width, df["Aswan_Visitors"],
        width=bar_width, color='skyblue', label="Aswan")

plt.title("Comparison of Visitors: Luxor & Aswan vs Egypt (20202025)")
plt.xlabel("Year")
plt.ylabel("Number of Visitors")
plt.xticks(x, years)    
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(6,6))
sizes = [df["Luxor_Visitors"].iloc[-1], df["Egypt_Visitors"].iloc[-1] - df["Luxor_Visitors"].iloc[-1]]
labels = ["Luxor", "Rest of Egypt"]
colors = ['orange', 'lightgray']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, explode=(0.1,0))
plt.title("Luxor's Share of Egypt's Tourism (2025)")
plt.show()
```

#percentage of Egypt Age Distribution

```python
import matplotlib.pyplot as plt

# Egypt age distribution data (using average values)
age_groups = [
    "Under 15 years",
    "15 - 29 years",
    "30 - 44 years",
    "45 - 59 years",
    "60 years and above"
]

population_percentages = [
    34,  # average of 33 - 35
    28,  # average of 27 - 29
    19,  # average of 18 - 20
    11,  # average of 10 - 12
    7    # average of 6 - 8
]

# Explode effect (same style as your chart)
explode = (0.05, 0.05, 0.05, 0.05, 0.05)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    population_percentages,
    labels=age_groups,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode
)

plt.title("Egypt Population Age Distribution")
plt.show()
```
